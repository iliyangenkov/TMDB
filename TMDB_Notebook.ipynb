```python
def extract_genres(x): # function provided by Alex
    ''' Function to extract genres for each datapoint. '''
    x = ast.literal_eval(x) ## Transform '['name', 'id']' back into ['name', 'id']
    Genres = [] ## Empty list to store the genres
    
    # Iterate through each dictionary
    for item in x: ## iterate for each dictionary in our list
        Genres.append(item['name']) ## Grab the 'name' key for each dictionary
        
    # Return the Genres
    return Genres

## Apply the function to genres
df['genres'] = df['genres'].apply(extract_genres)
```

```python
df.head(10)
```

```python
def extract_from_list(x): # repurpose the function to extract from the other columns
    ''' Function to extract genres for each datapoint. '''
    x = ast.literal_eval(x) ## Transform '['name', 'id']' back into ['name', 'id']
    variable = [] ## Empty list to store the genres
    
    # Iterate through each dictionary
    for item in x: ## iterate for each dictionary in our list
        variable.append(item['name']) ## Grab the 'name' key for each dictionary
        
    # Return the Genres
    return variable

## Apply the function to genres
df['keywords'] = df['keywords'].apply(extract_from_list)
```

```python
df['production_companies'] = df['production_companies'].apply(extract_from_list)
df['production_countries'] = df['production_countries'].apply(extract_from_list)
df['spoken_languages'] = df['spoken_languages'].apply(extract_from_list)
```

```python
# some of the extracted columns have empty lists. They should also be converted to null values
df[['genres', 'keywords','production_companies','production_countries','spoken_languages']] = df[['genres', 'keywords','production_companies','production_countries','spoken_languages']].apply(lambda col: col.apply(lambda x: np.nan if isinstance(x, list) and len(x) == 0 else x))
```

```python
# reassess null values
df.isnull().sum()
# homepage, keywords, overview, production_companies and tagline can be left as films could just not have them
# genres, release_date, runtime, spoken_languages, production_countries can simple be removed as they dont account for a large amount of entries
# budget and revenue cant be removed, and the will be estimated using the mean to maintain the distribution
```

```python
# dropped the null value rows
df.drop(df[df['genres'].isnull() == True].index, inplace = True)
df.drop(df[df['release_date'].isnull() == True].index, inplace = True)
df.drop(df[df['runtime'].isnull() == True].index, inplace = True)
df.drop(df[df['spoken_languages'].isnull() == True].index, inplace = True)
df.drop(df[df['production_countries'].isnull() == True].index, inplace = True)
```

```python
# filling the null values of revenue with the mean, which will maintain the distribution
df['revenue'] = df['revenue'].fillna(round(df['revenue'].mean(),-7))
```

```python
# do the same for budget
df['budget'] = df['budget'].fillna(round(df['budget'].mean(),-6))
```

## Data Exploration and Visualisation

### Movie Genre Distribution

```python
movie_genres = df.explode('genres').groupby('genres')['id'].count().sort_values(ascending = False).reset_index() # create an ordered dataframe of genres by number of films
movie_genres
```

```python
top_5 = movie_genres.head()
top_5 # top 5 are drama, comedy, thriller, action, romance
```

```python
bottom_5 = movie_genres.tail()
bottom_5 # bottom 5 are tv movies, foreign, western, documentary, war
```

```python
# Plot the Top 5 and Bottom 5 genres by Number of Films
sns.barplot(data = top_5,
            x = 'genres',
            y = 'id',
            palette = 'autumn'
           )
plt.xlabel('Genre')
plt.ylabel('Number of Films')
plt.title('Top 5 Genres by Number of Films')
plt.show()
```

```python
# Plot the Bottom 5 Genres by Number of Films
sns.barplot(data = bottom_5,
            x = 'genres',
            y = 'id',
            palette = 'winter'
           )
plt.xlabel('Genre')
plt.ylabel('Number of Films')
plt.title('Bottom 5 Genres by Number of Films')
plt.show()
```

<p> We can see which genres have the largest selection of films, and can be used for recommendations to customers who enjoy those genres, as well as new customers.</p>

<p> However, we also have genres that have a very small selection, and should get more curated films so that customers that enjoy them can have enough recommendations.</p>

#### Why is it that some genres have so many films, but others have so few?

Could it be Revenue?

```python
# create a table of genres by average revenue
revenue_genres = df.explode('genres').groupby('genres')['revenue'].mean().sort_values(ascending = False).reset_index()
revenue_genres
```

```python
# merge to the movie_genres table
movie_genres = pd.merge(movie_genres,revenue_genres, on = 'genres')
movie_genres
```

```python
# check the correlation between the number of films in a genre and average revenue of that genre
movie_genres[['id','revenue']].corr()
# basically no relationship between revenue and the number of films in that genre
```

Could it be popularily that drives the number of films in a genre?

```python
#create a table of genres by average popularity score
popularity_genres = df.explode('genres').groupby('genres')['popularity'].mean().sort_values(ascending = False).reset_index()
popularity_genres
```

```python
# merge to the movie_genres table
movie_genres = pd.merge(movie_genres,popularity_genres, on = 'genres')
movie_genres
```

```python
# check the correlation between the number of films in a genre and average popularity score of that genre
movie_genres[['id','popularity']].corr()
# there is a stronger relationship here. The popularity score of a genre does seem to be correlated with the number of films in that genre. This correlation is still relatively weak though.
```

What about budget? Perhaps a lower budget for films in a genre would allow for a larger number of films to be released?

```python
# create a table of genres by average budget
budget_genres = df.explode('genres').groupby('genres')['budget'].mean().sort_values(ascending = False).reset_index()
budget_genres
```

```python
# merge to movies table
movie_genres = pd.merge(movie_genres,budget_genres, on = 'genres')
movie_genres
```

```python
# check the correlation between the number of films in a genre and budget of that genre
movie_genres[['id','budget']].corr()
# also no correlation between the budget of a film and the number of films per genre
```

Finally, lets check the vote_average

```python
# create a table of genres by average vote_average
vote_genres = df.explode('genres').groupby('genres')['vote_average'].mean().sort_values(ascending = False).reset_index()
vote_genres
```

```python
# merge to movies table
movie_genres = pd.merge(movie_genres,vote_genres, on = 'genres')
movie_genres
```

```python
# check the correlation between the number of films in a genre and budget of that genre
movie_genres[['id','vote_average']].corr()
# a significant negative correlation
```

<p> The average vote for a genre actually seems to decrease the number of films in that genre.</p> 

<p> Why is that? A hypothesis is that the number of films in a genre is also correlated to the number of votes for that genre, which would lead to lower average votes, as average votes = sum of votes/vote count.</p>

```python
# create a table of genres by total vote_count
count_genres = df.explode('genres').groupby('genres')['vote_count'].sum().sort_values(ascending = False).reset_index()
count_genres
```

```python
# merge to movies table
movie_genres = pd.merge(movie_genres,count_genres, on = 'genres')
movie_genres
```

```python
# check the correlation between the number of films in a genre and total votes in that genre
movie_genres[['id','vote_count']].corr()
# a very strong positive correlation, which explains the aforementioned relationship
```

<p> It seems that the only significant driver for the number of films in a genre is the popularity of films in that genre. In that case, it seems that the genres that are most released are those that people enjoy, and those that underperform are released less. </p>

<p> However, its not that simple. </p>

```python
# plot the relationship between the number of films in a genre and the average popularity score of the genre
sns.scatterplot(data=movie_genres,
                x='popularity',
                y='id'
               )
plt.xlabel('Average Popularity Score of Genre')
plt.ylabel('Number of Films in Genre')
plt.title('Number of Films by the Average Popularity Score')
plt.show()
```

<p> As you can see from the above graph, there is only a slightly positive correlation. This still doesnt explain the large variation in the number of films across genres. Without further data provided, it would be hard to explain the distribution of the genres any further. However, the provided bar plots of the top 5 and bottom 5 genres can be used when making genre recommendations to new customers or choosing genres to curate more movies for, as the variation in film genres doesn't seem to be explained only by the popularity of the genres. </p>

<p> Another concern is that of sampling bias. It may simply be the case that this dataset mostly captures the characteristics of films in genres like drama, thriller etc, and not genres like documentaries or westerns. If that is the case, it is hard to make the inferences above as there isn't enough data to investigate the correlations. </p>

### Individual Film Analysis

<p> Are there films that are similar in one characteristic, but completely different in another? </p>

**Select two films that have high popularity, but one has a low budget, whereas the other has a high budget.**

```python
df.describe() # use quartiles to define high and low budget/popularity
```

```python
df[(df['budget'] >= 40000000) & (df['popularity'] >= 30)] # high budget, high popularity, Spectre
```

```python
df[(df['budget'] <= 12000000) & (df['popularity'] >= 30)] # low budget, high popularity, Psycho
```

The two films selected are Spectre, a high budget blockbuster, and Psycho, a much lower budget but equally popular horror film.

```python
pd.set_option('max_colwidth', None)
df[(df['title'] == 'Spectre') | (df['title'] == 'Psycho')] # compare both films
```

Similarities:
* Both are high popularity films

Differences:
* Wildly different budgets
* Very different genres
* Spectre has much larger revenue
* Different runtimes
* Psycho is better rated, with a comparable number of votes

#### These films are different in almost every aspect apart from popularity. Why are they equally popular? ####

<p> The context of these two films can explain are large amount of these differences. </p>

<p> Spectre is a modern installment in the very popular, long running Bond franchise. It's large budget is used for exotic locations and high quality special effects in the action scenes. It is immensely popular due to the brand and its high profile nature as a blockbuster. </p>

<p> Psycho was released much earlier in the 60s, and was a genre defining horror-slasher, that cultivated its audience through its unique (and at the time controversial) themes. Its low budget became a hallmark feature of the genre of horror slashers, which continue to be relatively low budget, but high popularity and high revenue. Its popularity is driven by its unique themes and its historical significance. </p>

<p> The impact that Psycho had at the time is what gives it such a high popularity score, as opposed to the high profile nature and brand of Spectre. </p>
